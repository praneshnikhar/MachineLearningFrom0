AP@k is calculated using P@k for example:
if wer have to calculate AP@3 , we calculate P@1, P@2, P@3 and the divier the sum by 3.

```python
def apk(y_true, y_pred, k):
    pk_values = []
    for i in range (1, k+1):
        pk_values.append(pk(y_true,y_pred, i))
    
    if len(pk_values) == 0:
        return 0
    return sum(pk_values)/len(pk_values)
```
