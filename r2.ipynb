```python
import numpy as np
def r2(y_true, y_pred):
    mean_true_value = np.mean(y_true)
    numerator = 0
    denominator = 0
    
    for yt,yp in zip(y_true, y_pred):
        numerator += (yt - yp)**2
        denominator += (yt - mean_true_value)**2
    ratio = numerator/denominator
    r2_score = 1 - ratio
    return r2_score
```
